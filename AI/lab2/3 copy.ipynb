```python
import math
import os
from collections import Counter

import pandas as pd
import matplotlib.pyplot as plt
from nltk.tokenize import word_tokenize, sent_tokenize
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
import numpy as np
from PIL import Image
```

```python
emp_data = pd.read_csv("data/employees.csv")
salaries = emp_data['Salary']
bonuses = emp_data['Bonus %']

scaler = MinMaxScaler()
salaries_scaled = scaler.fit_transform(pd.DataFrame(salaries))
bonuses_scaled = scaler.fit_transform(pd.DataFrame(bonuses))

fig, axs = plt.subplots(2, 2, figsize=(15, 10))
axs[0, 0].hist(salaries, 20, edgecolor='black', color='green')
axs[0, 0].set_title('Salary Original')
axs[0, 1].hist(salaries_scaled, 20, edgecolor='black', color='red')
axs[0, 1].set_title('Salary Normalizare')
axs[1, 0].hist(bonuses, 20, edgecolor='black', color='green')
axs[1, 0].set_title('Bonus Original')
axs[1, 1].hist(bonuses_scaled, 20, edgecolor='black', color='red')
axs[1, 1].set_title('Bonus Normalizat')

plt.tight_layout()
plt.show()
```

```python
img = Image.open('data/images/BERT.png').convert('L')
pixels = np.array(img)
pixels_normalized = pixels / 255.0

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(pixels.flatten(), bins=50, color='green')
plt.title('Histograma valorilor pixelilor originali')
plt.xlabel('Valoare pixel')
plt.ylabel('Frecvență')

plt.subplot(1, 2, 2)
plt.hist(pixels_normalized.flatten(), bins=50, color='red')
plt.title('Histograma valorilor pixelilor normalizați')
plt.xlabel('Valoare pixel')
plt.ylabel('Frecvență')

plt.tight_layout()
plt.show()
```
